$$

Определить функцию плотности $f_x(T)$.

```python
#C - diffusion
C = 0.1
#T - diffusion
T = 1.0
#n - trajectories
n = 50
#m - timesteps
m = 10
tau = T / m
```

Считаем траектории

```python
import numpy
t = [k * tau for k in range(m)]
x = [None] * n
for i in range(n):
    x[i] = [0.5] * m
for k in range(m - 1):
    for i in range(n):
        x[i][k + 1] = x[i][k] + C * tau**0.5 * numpy.random.normal()
```

Выводим траектории

```python
from matplotlib import pylab as plt
%matplotlib notebook

plt.figure(0)
for i in range(n):
    lines = plt.plot(t, x[i])
```

Считаем функцию плотности

```python
def make_hist(x_list):
    H = 8
    hist = [0] * H
    for x in x_list:
        if (x >= 0) and (x <= 1):
            index = int(x * H)
            hist[index] += 1
    return hist

x_last = [x[i][m - 1] for i in range(n)]
hist = make_hist(x_last)
```

Выводим функцию плотности

```python
plt.figure(1)
lines = plt.plot(hist)
```
